````python
# Import the necessary module
import pandas as pd
import re

# Load the tracks table
tracks = pd.read_csv('../Cours1/datasets/tracks.csv')

# Define the split_string function
def split_string(x):
    return re.findall(r"'([^']*)'", x)

# Replace the id_artists column in tracks with a split version
tracks['id_artists'] = tracks['id_artists'].apply(split_string)

# Explode the id_artists column to get one row per artist
tracks = tracks.explode('id_artists')
````

````python

# Merge the tracks and artists tables on the id_artists column
df = pd.merge(tracks, artists, how="left")
````

````python
df
````

````python
# Votre code
# Drop unnecessary columns
df = df.drop(['id_artists', 'id_x','id_y'], axis=1)

````

````python
# Rename columns
df = df.rename(columns={'name_x': 'track_name', 'name_y': 'artist_name', 'followers': 'artist_followers'})
````

##  ⚒️ Commandes Chipotle

1. Importer le fichier `chipotle.tsv` dans pandas.
2. Interpréter et potentiellment nettoyer les données du tableau.
3. Combien de produits coûtent plus de $10 par unité ?
4. Lister le prix moyen de chaque item (soyez prudent)
5. Afficher la commande la plus chère.
6. Combien de fois le 'veggie bowl' a-t-il été commandé ?
7. Combien de fois quelqu'un a-t-il commandé plus d'un soda ?

````python
# Votre code

chipotle = pd.read_csv('../Cours2/datasets/chipotle.tsv', sep='\t')
````

````python
chipotle.info()
````

````python
chipotle['item_price'] = chipotle['item_price'].apply(lambda x: float(x[1:]))
````

````python
# How many products cost more than $10 per unit?
num_expensive = (chipotle['item_price'] / chipotle['quantity'] > 10).sum()
print(f'{num_expensive} products cost more than $10 per unit.')
chipotle["unit_price"] = chipotle["item_price"] / chipotle["quantity"]
# Calculate the mean price of each item
item_means = chipotle.groupby('item_name')['unit_price'].mean()
print(item_means)

# What is the most expensive order?
most_expensive = chipotle.groupby('order_id')['item_price'].sum().nlargest(1)
print(f'The most expensive order was {most_expensive.values[0]:.2f} dollars.')

# How many times was the 'veggie bowl' ordered?
num_veggie_bowl = (chipotle['item_name'] == 'Veggie Bowl').sum()
print(f'The Veggie Bowl was ordered {num_veggie_bowl} times.')

# How many times did someone order more than one soda?
num_multiple_sodas = ((chipotle['item_name'] == 'Canned Soda') | (chipotle['item_name'] == 'Canned Soft Drink')) & (chipotle['quantity'] > 1)
print(f'{num_multiple_sodas.sum()} orders included more than one soda.')
````

## ⚒️ Films IMDB

1. Importer le fichier `movies.csv` dans pandas.
2. Interpréter et potentiellment nettoyer les données du tableau.
3. Quel est le film le plus long ? Quelle est la durée moyenne d'un film par genre ?
4. Compter le nombre de film par catégorie de rating.
5. Déterminer le film le mieux noté par catégorie.
6. Lister les films ayant le même titre.
7. Lister le nombre de fois où chaque acteur apparaît dans un film. Quels sont les 10 acteurs apparaissant le plus souvent ?

````python
# Votre code


movies = pd.read_csv('./datasets/movies.csv')

# Print the first few rows of the dataframe
print(movies.head())

# Check for missing values
print(movies.isna().sum())

# Drop the missing values
movies = movies.dropna()

# What is the longest movie?
longest_movie = movies.loc[movies['duration'].idxmax()]
print(f'The longest movie is {longest_movie["title"]} with a duration of {longest_movie["duration"]} minutes.')

# What is the average duration of a movie by genre?
genre_duration = movies.groupby('genre')['duration'].mean()
print(genre_duration)


````

````python
movies.columns
````

````python
# How many movies are there in each rating category?
rating_counts = movies['content_rating'].value_counts()
print(rating_counts)
# What is the highest rated movie in each rating category?
highest_rated = movies.groupby('content_rating').apply(lambda x: x.loc[x['star_rating'].idxmax()])
print(highest_rated[['title', 'content_rating', 'star_rating']])

# Which movies have the same title?
same_title = movies[movies.duplicated(['title'], keep=False)]
print(same_title[['title']])

# How many times does each actor appear in a movie?
actor_counts = movies['actors_list'].str.split('|', expand=True).stack().value_counts()
print(actor_counts.head(10))
````
