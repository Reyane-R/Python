### 3.2.5. Joins

Nous allons maintenant parler d'une opération courante en pandas et SQL: le join de deux tables (i.e. mettre en commun les données de plusieurs tables selon une variable commune).

````python
import pandas as pd

# Create DataFrames for Orders and Customers
orders_data = {
    'OrderID': [1, 2, 3, 4],
    'CustomerID': [101, 102, 101, 104],
    'OrderDate': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-03-15']
}

customers_data = {
    'CustomerID': [101, 102, 103, 101],
    'CustomerName': ['Alice', 'Bob', 'Charlie', 'Alice'],
    'ContactInfo': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 'alice2@email.com']
}


orders_df = pd.DataFrame(orders_data)
customers_df = pd.DataFrame(customers_data)

print("\nLEFT TABLE")
display(orders_df)
print("\nRIGHT TABLE")
display(customers_df)

# INNER JOIN equivalent in Pandas
inner_join_result = pd.merge(orders_df, customers_df, on='CustomerID', how='inner')

# LEFT JOIN (OUTER JOIN) equivalent in Pandas
left_join_result = pd.merge(orders_df, customers_df, on='CustomerID', how='left')

# RIGHT JOIN (OUTER JOIN) equivalent in Pandas
right_join_result = pd.merge(orders_df, customers_df, on='CustomerID', how='right')

# FULL OUTER JOIN (OUTER JOIN) equivalent in Pandas
full_outer_join_result = pd.merge(orders_df, customers_df, on='CustomerID', how='outer')

# Print the results
print("\nINNER JOIN Result: keep only the intersection !")
display(inner_join_result)

print("\nLEFT JOIN Result: order is the master ! ")
display(left_join_result)

print("\nRIGHT JOIN Result: customer is the master ! ")
display(right_join_result)

print("\nFULL OUTER JOIN Result: we take all from both the table ")
full_outer_join_result
````
