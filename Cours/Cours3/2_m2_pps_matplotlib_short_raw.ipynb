# Matplotlib

Matplotlib est une librairie de visualisation en Python permettant de générer plusieurs types de graphes: histogrammes, nuages de points etc.

## 4.1. Utilisation de Matplotlib

### 4.1.1 Elements de base

Matplotlib utilise plusieurs approches pour la génération de graphe. Nous allons préférer l'approche orientée objet. Cette approche définit deux éléments: la `figure` (le contenant) et les `axes` (le contenu). La création de la figure elle-même se fait avec la commande plot, suivi de deux arguments: les `x` et `y`.

```python
import matplotlib.pyplot as plt
import numpy as np
# Exemple: plot approche objet (fig, ax)
fig, ax = plt.subplots()

x = np.linspace(0, 10, num=1000)
y = np.cos(x)

ax.plot(x, y)
```

### 4.1.2. Backends et Sauvegarde

Matplotlib peut utiliser plusieurs backends pour générer et afficher une figure. Par exemple, le snippet ci-dessous affiche la figure dans un navigateur web (`matplotlib.use('webagg')`):

```python
# Exemple: Backend web - À utiliser au début du script
import matplotlib
matplotlib.use('webagg')

fig, ax = plt.subplots()
x = np.linspace(0, 10, num=1000)
y = np.cos(x)
ax.plot(x, y)

plt.show()
```

Il est possible de sauvegarder sur disque une figure générée avec matplotlib avec la commande `savefig`.

```python
# Exemple: Utilisation de savefig
fig.savefig("figure1.png")

# for i in range(1, 11):
#     fig.savefig(f"figure{i}.png")
```

## 4.2. Style et attributs

### 4.2.1. Style et taille de la figure

Les attributs et configuration des figures matplotlib sont directement modifiés par des appels à des fonction Python. Par exemple, il est possible de définir le style général d'une figure matplotlib (`default`, `classic`, `bmh`, `dark_background`, `ggplot`).

```python
# Exemple: Style d'une figure (plt.style.use('classic'))
# plt.style.use("dark_background")
plt.style.use("default")

fig, ax = plt.subplots()
x = np.linspace(0, 10, num=1000)
y = np.cos(x)
z = np.sin(x)
ax.plot(x, y)
ax.plot(x, z)
```



Il est possible de spécifier directement un grand nombre d'attributs lors de la création d'une figure. Par exemple, la taille d'une figure en pixels (x100)

```python
# Exemple: Taille d'une figure (figsize=)
fig, ax = plt.subplots(figsize=(10, 10))
x = np.linspace(0, 10, num=1000)
y = np.cos(x)
z = np.sin(x)
ax.plot(x, y)
ax.plot(x, z)
```

### 4.2.2. Axes

Autre élément essentiel d'une figure: ses axes. Il existe un grand nombre de fonctions permettant de définir ou modifier les axes x ou y d'une figure.

```python
# Exemple: Types d'axes (set_xscale: linear, log, function)

x = np.linspace(0, 100, num=10000)
# y = np.exp(x**(1/2))
y = np.exp(x)
# y = np.exp(x + np.random.randint(0, 100))

fig, ax = plt.subplots()
ax.set_yscale("log")
# ax.set_xscale("log")
ax.plot(x, y)
```

Pour les figures dont on veut contrôler les limites, les options `xlim` et `ylim` sont disponibles.

```python
# Exemple: Utilisation de ylim et xlim. 
# Egalement plt.axis pour tout faire d'un coup. Egalement, 'tight', 'equal'

fig, ax = plt.subplots()

x = np.linspace(0.1, 10, num=1000)
ax.plot(x, np.cos(x), label="$cos(x)$")
ax.plot(x, np.sin(x), label="$sin(x)$")
ax.plot(x, x, label="$x$")
ax.plot(x, x**(1/2), label="$\sqrt{x}$")
ax.plot(x, np.log(x), label="$\log{x}$")
ax.plot(x, np.exp(x), label="$e^x$")

ax.axhline(y=5, linestyle="--", color="grey")

ax.set_xlim(0, 10)
ax.set_ylim(-2, 10)

ax.legend()
```

### 4.2.3. Labels

Très important lorsqu'on présente une courbe ou un graphe il faut indiquer son titre et ses axes avec l'unité correspondante. À utliser avec `title`, `xlabel`, `ylabel`.

```python
# Exemple: title, xlabel, ylabel Montrer différence avec ax.set_xlabel/lim/title

fig, ax = plt.subplots()

x = np.linspace(0.1, 10, num=1000)
ax.plot(x, np.cos(x), label="$cos(x)$")
ax.plot(x, np.sin(x), label="$sin(x)$")
ax.plot(x, x, label="$x$")
ax.plot(x, x**(1/2), label="$\sqrt{x}$")
ax.plot(x, np.log(x), label="$\log{x}$")
ax.plot(x, np.exp(x), label="$e^x$")

ax.axhline(y=5, linestyle="--", color="grey")

ax.set_xlim(0, 10)
ax.set_ylim(-2, 10)

ax.set_title("Comparaison de plusieurs fonctions")
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$")

ax.legend()
```

## 4.2. Type de figures

Une fois les bases maîtrisées, il faut également apprendre à maîtriser les différents types de visualisations en fonction des données à représenter.

Pour illustrer chacun de ces types, nous allons utiliser un nouveau dataset: la base de donnée de tous les joueurs de football sur FIFA 21.

```python
import pandas as pd
X = pd.read_csv("./datasets/fifa.csv")
```

```python
pd.set_option('display.max_columns', None)
X.head()
```

```python
# Petits rappels: lister le nombre de joueurs de chaque pays dans ce dataset.
X.groupby("nationality")["short_name"].count().sort_values(ascending=False)
```

### 4.2.1. Plots

Le type basique de figures, déjà vu jusqu'ici. Ce type de graphe est en général utilisé pour représenter l'évolution d'une variable dans le temps (i.e. cours d'une action en bourse en fonction du temps, etc).

```python
# Exemple: Courbe exponentielle
fig, ax = plt.subplots()

x = np.linspace(0.1, 10, num=1000)
ax.plot(x, np.exp(x), label="$e^x$");
```

### 4.2.2. Bar charts

Les bar charts sont utilisés pour comparer une valeur entre plusieurs catégories.

```python
# Exemple: Bar chart synthétique
fig, ax = plt.subplots(figsize=(18, 3))
# ax.bar(["a", "b", "c"], [3, 4, 5])

Z = X.groupby("nationality")["short_name"].count().sort_values(ascending=False)[:10]
ax.bar(Z.index, Z.values);
```

### 4.2.3. Histplots

Les histogrammes permettent de représenter la distribution d'une variable sur un nombre d'observations données.

```python
# Exemple: Hist chart de la répartition du poids de tous les joueurs de la base.
# Quels sont les 5 joueurs les plus lourds ? Les moins lourds ?
fig, ax = plt.subplots()
# help(ax.hist)

ax.hist(X.weight_kg, bins=35);
```

### 4.2.4. Pie Charts

Les pie charts sont généralement utilisés pour représenter les proportions d'une catégorie dans un ensemble donné.

```python
# Exemple: Proportion des joueurs par pays
Z = X.groupby("nationality").sofifa_id.count().sort_values(ascending=False)[:10]

fig, ax = plt.subplots(figsize=(7,7))
ax.pie(Z.values, labels=Z.index);
```

### 4.2.5. Scatterplots

Un type de graphe plus avancé qui peut représenter les correlations entre deux, trois ou quatre variables, sur un échantillon donné.

```python
# Exemple: Corrélation poids et taille (utiliser iterrows pour label avec ax.annotate)

fig, ax = plt.subplots(figsize=(20,10))
ax.scatter(X.weight_kg[:100], X.height_cm[:100], s=X.wage_eur[:100]/1000, c=X[:100].team_position.astype('category').cat.codes)

for _, val in X[:100].iterrows():
    ax.annotate(val.short_name, (val.weight_kg, val.height_cm))
```
